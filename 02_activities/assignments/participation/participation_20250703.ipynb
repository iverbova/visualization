````python
#  Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
````

- `NumPy`, `Pandas`, and `SciPy` provide numerical, tabular, and scientific computation tools.
- `Matplotlib` is used for all visualizations and style customization.


````python
# Generate sample data to use in multiple plots
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
````

Generated two sets of random values (`y1` and `y2`) over a sequence of 50 values (`x`). These will be used to demonstrate various styling and labeling features in plots.


````python
# Basic line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
````

````python
# Basic line plot with legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc='lower right')
plt.title("Line Plot with Legend")
plt.show()
````

Added a legend using `label` and `ax.legend()` to distinguish the two lines. This improves clarity for multi-series plots.


````python
# Customize legend appearance
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc='lower right', frameon=True, fontsize=12, ncol=2, shadow=True)
plt.title("Styled Legend")
plt.show()
````

````python
# Position legend outside of plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label="Person 1")
ax.plot(x, y2, label="Person 2")
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title("Legend Positioned Outside")
plt.show()
````

Using `bbox_to_anchor`, we can move the legend outside the plot area to save space and reduce visual clutter.


````python
# Add text annotations to highlight values
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.text(10, 95, "This value is important!", ha='center', color='red', size=8)
plt.title("Text Annotation")
plt.show()
````

````python
# Add annotation with an arrow
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.annotate("Look here!", xy=(10, 95), size=8, ha='center', xytext=(20, 90),
            arrowprops=dict(arrowstyle="->", color="green"))
plt.title("Arrow Annotation")
plt.show()
````

Annotations with arrows help direct attention to specific data points, supporting interpretability.


````python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])
````


Using `ax.axis([0, 10, 0, 10])` sets fixed limits for the x- and y-axes. This creates a defined plotting area from 0 to 10 on both axes, which is helpful for precise placement of elements like text, shapes, or annotations â€” especially when working with layout-based designs.


````python
fig, ax = plt.subplots()
ax.axis([0,10,0,10]) # Set axis range from 0 to 10 on both x and y

# Place text using different coordinate systems
ax.text(1, 5, ". Data:(1, 5)",           transform=ax.transData)     # Data coordinates
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)",   transform=ax.transAxes)     # Relative to axes
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure)   # Relative to full figure
````

This example demonstrates how `matplotlib` allows placing text based on different coordinate systems:

- `transData`: Positions text using actual data values (e.g., `(1, 5)` refers to the data point itself).
- `transAxes`: Places text relative to the axis (ranges from 0 to 1 across the axis area).
- `transFigure`: Places text relative to the entire figure window, useful for global annotations like footnotes or titles.

This flexibility is useful for placing labels in ways that remain consistent regardless of how the plot is scaled or zoomed.

````python
fig, ax = plt.subplots()
ax.plot(x, y1)

# Data coordinates: attaches to a specific data point
ax.text(x=10, y=y1[10], s="Data point (10, y1)", color="blue", transform=ax.transData)

# Axes coordinates: relative to the axes [0, 1], independent of data
ax.text(0.7, 0.9, "Axes (0.7, 0.9)", color="green", transform=ax.transAxes)

# Figure coordinates
ax.text(0.15, 0.6, "Figure 0.15, 0.6", color="purple", transform=fig.transFigure)

plt.title("Annotations Using Different Coordinate Systems")
plt.show()
````

````python
# Annotating a scatter plot with a basic arrow

fig, ax = plt.subplots(figsize=(5, 3))

# Scatter plot for two sets of points
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")

# Add legend in the upper right corner
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Annotate a specific data point with an arrow and label
ax.annotate('This is important!',
            xy=(10, 95),            # location the arrow points to
            xytext=(20, 94),        # location of the text label
            arrowprops=dict(facecolor='black'))  # simple black arrow

plt.title("Basic Arrow Annotation")
plt.show()
````

````python
# Styled arrow annotation using arrowprops

fig, ax = plt.subplots(figsize=(5, 3))

# Same scatter setup
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Annotate with custom arrow style and color
ax.annotate('This is important!',
            xy=(10, 95),
            xytext=(20, 94),
            arrowprops=dict(arrowstyle="wedge", color="hotpink"))

plt.title("Custom Arrow Style")
plt.show()
````

````python
# Customize axis ticks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

ax.yaxis.set_major_locator(plt.NullLocator())  # Remove y-axis ticks
ax.xaxis.set_major_formatter(plt.NullFormatter())  # Remove x-axis labels
plt.title("Axis Ticks Removed")
plt.show()
````

````python
# Limiting tick marks 

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

ax.yaxis.set_major_locator(plt.MaxNLocator(3))  # display at most 3 nicely spaced ticks on the y-axis

plt.title("Y-Axis Tick Reduced")
plt.show()
````

````python
# Setting fixed intervals for tick marks on the x-axis
from matplotlib.ticker import MultipleLocator
fig, ax = plt.subplots(figsize=(5, 3))

# Scatter plot with two groups
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")

# Add legend for clarity
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Set major ticks on x-axis to show every 5 units
ax.xaxis.set_major_locator(MultipleLocator(5))

plt.title("Tick Marks Every 5 Units")
plt.show()
````

To control the spacing of tick marks explicitly, use `MultipleLocator`.

For example:
```python
ax.xaxis.set_major_locator(MultipleLocator(5))
```
will place a tick on every 5th unit along the x-axis.

This helps improve clarity or consistency across multiple plots.


````python
# Rotating x-axis labels for better fit or clarity

fig, ax = plt.subplots(figsize=(5, 3))

# Basic scatter plot with overlapping categories
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Rotate the x-axis tick labels for better readability
plt.xticks(rotation=45, ha='right')  # ha: horizontal alignment

plt.title("Rotated Axis Labels")
plt.show()
````

Sometimes, tick labels overlap or become hard to read â€” especially with long or crowded values.

You can rotate them using:
```python
plt.xticks(rotation=45, ha='right')
```

- `rotation=45` tilts the labels diagonally.
- `ha='right'` aligns the text so it doesn't get clipped.

This improves legibility and layout, especially when the x-axis contains many values.


### Customizing Axis Title Fonts

In this activity, the focus is on enhancing plot readability and aesthetics by modifying the axis title. The task involves applying custom font properties (such as color and typeface) to the x-axis label using `plt.xlabel()` and the `fontdict` parameter. This builds on earlier lessons related to styling and layout customization in Matplotlib.


````python
# Modify axis label style using fontdict
font1 = {'family': 'serif', 'color': 'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1)
plt.xlabel("Shiny New X Axis!", fontsize=12, fontdict=font1)


# Change tick label font properties
for label in ax.get_xticklabels():
    label.set_fontfamily('serif')
    label.set_color('indigo')

for label in ax.get_yticklabels():
    label.set_fontfamily('serif')
    label.set_color('indigo')



plt.title("Custom Font Axis Label")
plt.show()
````

Understanding Styles in Matplotlib
Matplotlib offers pre-defined styles that help apply a consistent aesthetic across visualizations. Styles change elements like colors, fonts, and backgrounds. You can explore the available styles and apply one before plotting your figures.



````python
# Check what built-in styles are available in matplotlib
import matplotlib.pyplot as plt

plt.style.available
````

````python
# Create sample data for plotting
np.random.seed(613)  # For reproducibility
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
````

### Attempt Using "ggplot" Style

The "ggplot" style mimics the popular aesthetics of R's ggplot2 library â€” with a gray background, gridlines, and clean color palette.

````python
# Apply ggplot style before plotting
plt.style.use('ggplot')

# Create the same figure with ggplot styling
fig, ax = plt.subplots(figsize=(5, 3))

# Plot both data series
ax.plot(x, y1, label="Line 1")
ax.plot(x, y2, label="Line 2")

# Add legend for clarity
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
# Show the plot
fig.show()
````

### Attempt Using `_mpl-gallery-nogrid` style
This style disables grid lines and offers a clean, minimal aesthetic often used in gallery-style visualizations. Itâ€™s useful when a less cluttered view is desired.

````python
# Apply the _mpl-gallery-nogrid style
plt.style.use('_mpl-gallery-nogrid')

# Create a figure and axis with a consistent size
fig, ax = plt.subplots(figsize=(5, 3))

# Plot two random series
ax.plot(x, y1, label="Series A")
ax.plot(x, y2, label="Series B")

# Add legend to distinguish lines
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))


plt.title("Minimal Look with '_mpl-gallery-nogrid' Style")

# Display the plot
plt.show()

````

### Attempt Using "classic" Style
The "classic" style brings back the original default matplotlib appearance, with simpler design and no gridlines.

````python
# Revert to classic style
plt.style.use('classic')

# Plot using the classic appearance
fig, ax = plt.subplots(figsize=(5, 3))

# Draw two data lines
ax.plot(x, y1, label="Line 1")
ax.plot(x, y2, label="Line 2")

# Add legend
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the result
fig.show()
````
