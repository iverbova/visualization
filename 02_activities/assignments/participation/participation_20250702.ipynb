```python
# Library Imports

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Set a seed for reproducibility and generate sample data
np.random.seed(613)
x = np.arange(50)  # x-axis: sequence of 50 values
y = np.random.randint(0, 100, 50)  # y-axis: 50 random integers from 0 to 100
```

Here, `np.arange(50)` creates a sequence from 0 to 49, and `np.random.randint(0, 100, 50)` generates 50 random integers. This data will be used in various plot types.


```python
# Create a scatter plot using matplotlib
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
plt.show()
```

A scatter plot shows individual data points, useful for identifying patterns, clusters, or outliers.


```python
# Create a bar chart with the same data
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)
plt.show()
```

Bar charts help compare quantities across categories or time. Each bar height corresponds to a value in `y`.


```python
# Create a line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
plt.show()
```

Line plots show trends over continuous intervals â€” commonly used for time-series or progression-based data.


```python
# Create a histogram of the y-values
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
plt.show()
```

Histograms display the distribution of values â€” in this case, showing how the 50 random integers are spread across bins.


```python
# Adding labels and title to a plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
plt.show()
```

Axes titles and labels improve readability and convey context for the data.


```python
# Define custom font styles
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# Apply custom fonts to plot text
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
plt.show()
```

Using `fontdict`, plot text can be styled for emphasis or accessibility.


```python
# Move the title to the left
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict=font1, loc='left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
plt.show()
```

```python
# Customize a scatter plot's appearance
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker='*', color='indigo')
plt.show()
```

This version uses a star marker and changes the color of points to indigo for visual emphasis.


```python
# Customize line appearance
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*', color='indigo', linestyle='--', linewidth=2)
plt.show()
```

```python
# Add hex color and style markers
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*',
        markersize=12,
        color='#7425b9',
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#fa9359')
plt.show()
```

```python
# Add and style gridlines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.grid(axis='y', color='blue', linewidth=2, linestyle='-.')
plt.show()
```

Gridlines help align values visually. They can be customized for better readability.

 **Activity: Reproducing a Chart from the Python Graph Gallery**
In this exercise, the goal was to explore a visualization example from the Python Graph Gallery. The contour plot using Seaborn was selected for its unique representation of density and gradient through filled contours.

Steps Followed:

The selected example is a **contour plot** using Seaborn, which visualizes the density of points in a 2D space using smooth gradients. - chose this chart because it introduces a different visual style (contour plot) and uses seaborn, which wasnâ€™t part of the main lesson.

```python
# Original contour plot from the Python Graph Gallery
import seaborn as sns

# Dataset 
df = sns.load_dataset('iris')
```

```python
# Plot using seaborn
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length)
plt.title("Original Contour Plot from Gallery")
plt.show()
```

This plot uses kernel density estimation (KDE) to visualize the density of points in the Iris dataset, using sepal width and sepal length as axes. It uses default colors and contour lines only (no fill). To improve visual clarity, a filled version with a red gradient was created. 


```python
# Change colors
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length, cmap="Reds", fill=True)
plt.title("Adjusted Contour Plot from Gallery")
plt.show()
```

Filling the contours improves the readability of the plot and makes the gradient of density easier to interpret. The red color map adds visual emphasis.


```python
# Change smoothing parameter
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length,
            cmap="Blues", fill=True, bw_adjust=0.5)
plt.title("Adjusted Contour Plot from Gallery")
plt.show()
```

Reducing the `bw_adjust` value makes the contours tighter and the plot more sensitive to small clusters or variations in the data. This increases detail but can also introduce noise. Smoothing was adjusted to highlight fine variations in density. This shows how visualization choices can impact the story the data tells.


### Evaluation Summary

- **Aesthetic**: The filled versions, especially with custom color maps, are more visually appealing and easier to interpret than the original outline-only plot.
- **Substantive**: All versions accurately reflect density patterns in the Iris dataset, though bandwidth adjustment impacts detail.
- **Perceptual**: The filled plots enhance clarity, and the color maps help guide the viewer's attention to areas of high and low density.


**Customized Contour Plot**

To go beyond the original gallery example, a final version of the contour plot was created with additional parameter tuning. These adjustments were made to enhance visual clarity and improve perceptual understanding of the density distribution.


```python
# Set visual style
sns.set_style("white")

# Create a filled contour plot with custom settings
# - 'cmap' changes the color map to a blue gradient
# - 'fill=True' fills the contour areas
# - 'bw_adjust' makes the contours tighter and the plot more sensitive to variations in data
# - 'alpha=0.7' adjusts transparency
# - 'levels=10' increases the number of contour layers
sns.kdeplot(x=df.sepal_width, y=df.sepal_length,
            cmap="Blues", fill=True, bw_adjust=0.5, alpha=0.7, levels=10)

# Add axis labels with larger font for clarity
plt.xlabel("sepal_width", fontsize=12)
plt.ylabel("sepal_length", fontsize=12)

# Add a clear title
plt.title("Adjusted Contour Plot from Gallery")

plt.show()
```

This final version uses additional parameters to improve the aesthetic and interpretive quality of the chart:
- `fill=True` and `alpha=0.7` make density areas easier to distinguish without being visually overwhelming.
- `levels=10` increases detail and helps identify finer gradients in the data.
- Font sizes were adjusted for readability.

These changes highlight how small adjustments in visualization parameters can greatly improve how data is perceived and understood.
